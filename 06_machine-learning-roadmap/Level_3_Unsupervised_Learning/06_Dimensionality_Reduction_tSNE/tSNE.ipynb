# Part 3.6: Unsupervised Learning - t-SNE

**t-Distributed Stochastic Neighbor Embedding (t-SNE)** is a non-linear dimensionality reduction technique used primarily for visualizing high-dimensional datasets. Unlike PCA, which tries to preserve global variance, t-SNE focuses on preserving the local structure of the data, meaning it tries to keep similar data points close to each other in the lower-dimensional map.

### Key Points about t-SNE
- **It's for visualization**: The resulting plot is a visualization, not a set of new features for modeling. The axes have no direct interpretation.
- **It's computationally expensive**: It does not scale well to very large datasets.
- **Hyperparameters matter**: The `perplexity` parameter, related to the number of nearest neighbors considered for each point, can significantly change the resulting plot.

```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits

# Load the digits dataset
digits = load_digits()
X = digits.data
y = digits.target
```

### Applying t-SNE

```python
# It's recommended to use PCA to reduce dimensions to a reasonable number first
# before applying t-SNE, but for this small dataset, we can apply it directly.

tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X)

print(f"Shape after t-SNE: {X_tsne.shape}")
```

### Visualizing the t-SNE Plot
The resulting visualization often shows much clearer separation between clusters than PCA.

```python
plt.figure(figsize=(10, 8))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', edgecolor='none', alpha=0.7)
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('t-SNE Visualization of Digits Dataset')
plt.colorbar()
plt.show()
```
