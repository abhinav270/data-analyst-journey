# Part 3.4: Unsupervised Learning - Gaussian Mixture Models (GMM)

A Gaussian Mixture Model (GMM) is a probabilistic model that assumes all the data points are generated from a mixture of a finite number of Gaussian distributions with unknown parameters. Unlike k-Means, which performs 'hard' clustering (assigning each point to a single cluster), GMMs perform 'soft' clustering. They provide the probability that a data point belongs to each of the clusters.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs

# Generate data with elongated clusters, which is hard for k-Means
X, y_true = make_blobs(n_samples=400, centers=4, cluster_std=0.60, random_state=0)
rng = np.random.RandomState(13)
X_stretched = np.dot(X, rng.randn(2, 2))
```

### Training a GMM

```python
# n_components is the number of clusters (Gaussian distributions)
gmm = GaussianMixture(n_components=4, random_state=42)
gmm.fit(X_stretched)
labels = gmm.predict(X_stretched)
```

### Visualizing the Clusters
GMM is powerful because it can model clusters of different shapes and sizes (ellipsoids).

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_stretched[:, 0], X_stretched[:, 1], c=labels, s=40, cmap='viridis', zorder=2)
plt.title('Gaussian Mixture Model Clustering')
plt.show()
```

### Soft Clustering
We can also look at the probabilities of each point belonging to each cluster.

```python
probs = gmm.predict_proba(X_stretched)
print("Probabilities for the first 5 points:")
print(probs[:5].round(3))
```
