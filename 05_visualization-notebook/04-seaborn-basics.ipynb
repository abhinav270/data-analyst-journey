# Seaborn Part 1: The Basics

Seaborn is a high-level statistical data visualization library built on top of Matplotlib. It provides a more attractive and convenient interface for creating common statistical plots. This notebook covers the basics of setting up Seaborn and creating simple plots.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

### Default Themes and Palettes

One of Seaborn's biggest advantages is its beautiful default styles. You can easily set the theme for all your plots.

```python
sns.set_theme(style="whitegrid", palette="viridis")

# Now, even a standard Matplotlib plot will have the Seaborn style
plt.plot([1, 2, 3, 4])
plt.title("Matplotlib plot with Seaborn theme")
plt.show()
```

### Understanding Seaborn's Data Structure Requirements

Seaborn works best with **long-form** (or 'tidy') data. This means:
1. Each variable forms a column.
2. Each observation forms a row.

Let's create a sample DataFrame in this format.

```python
data = {
    'day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'group': ['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B'],
    'value': [10, 12, 15, 13, 18, 14, 16, 11, 19, 20]
}
df = pd.DataFrame(data)
df
```

### Basic Plots

#### `scatterplot`

```python
# Load one of Seaborn's built-in datasets
tips = sns.load_dataset("tips")

sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
plt.title('Tip Amount vs. Total Bill')
plt.show()
```

#### `lineplot`

```python
# Using our custom long-form data
sns.lineplot(data=df, x='day', y='value', hue='group', style='group', markers=True)
plt.title('Value over Time by Group')
plt.show()
```

#### `barplot`
A barplot shows an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars.

```python
sns.barplot(data=tips, x="day", y="total_bill", hue="sex")
plt.title('Average Total Bill by Day and Sex')
plt.show()
```

#### `countplot`
A countplot is like a histogram for categorical variables. It shows the counts of observations in each categorical bin.

```python
sns.countplot(data=tips, x="day", hue="smoker")
plt.title('Count of Smokers and Non-Smokers by Day')
plt.show()
```
