### Pyplot vs. Object-Oriented API

Matplotlib has two main interfaces:
1.  **Pyplot Interface (State-based):** A collection of functions in `matplotlib.pyplot` that operate on a single, global figure. It's great for quick, simple plots.
2.  **Object-Oriented (OO) API:** The recommended approach for complex plots. You create `Figure` and `Axes` objects and call methods on them. It gives you more control and is better for writing reusable plotting functions.

We will primarily use the **Object-Oriented API** as it's more powerful and flexible.

```python
# OO API Example
fig, ax = plt.subplots()  # Create a figure and a set of subplots (in this case, just one)
ax.plot([1, 2, 3, 4], [10, 20, 25, 30])
plt.show()
```

### Creating Simple Plots

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

**Line Plot**

```python
fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

**Scatter Plot**

```python
x_rand = np.random.rand(50)
y_rand = np.random.rand(50)

fig, ax = plt.subplots()
ax.scatter(x_rand, y_rand)
plt.show()
```

**Bar Chart**

```python
categories = ['A', 'B', 'C']
values = [10, 25, 15]

fig, ax = plt.subplots()
ax.bar(categories, values)
plt.show()
```

**Histogram**

```python
data = np.random.randn(1000)

fig, ax = plt.subplots()
ax.hist(data, bins=30)
plt.show()
```

### Customizing Your Plot

#### Figure Size & Resolution

```python
# figsize=(width, height) in inches, dpi = dots per inch
fig, ax = plt.subplots(figsize=(8, 4), dpi=100)
ax.plot(x, y)
plt.show()
```

#### Titles, Labels, and Legends

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), label='sin(x)')
ax.plot(x, np.cos(x), label='cos(x)')

ax.set_title("Trigonometric Functions")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.legend()

plt.show()
```

#### Line Styles, Markers, and Colors

```python
fig, ax = plt.subplots()
ax.plot(x, x, color='red', linestyle='--', marker='o', label='y=x')
ax.plot(x, x+1, color='#00FF00', linestyle=':', marker='x', label='y=x+1') # Hex color
ax.legend()
plt.show()
```

#### Axis Limits and Scales

```python
fig, ax = plt.subplots()
ax.plot(x, np.exp(x))

# Set axis limits
ax.set_xlim(0, 10)
ax.set_ylim(0, 20000)

# Set scale to logarithmic
ax.set_yscale('log')

ax.set_title("Exponential Function with Log Scale")
plt.show()
```
